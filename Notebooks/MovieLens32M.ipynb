```python
movies.columns = ['movieId', 'title', 'genres']

movieId	é categórico nominal,	é o identificador  do filme.
title	é categórico nominal,	nome do filme.
genres	é categórico nominal,	lista de gêneros
```

```python
ratings.columns = ['userId', 'movieId', 'rating', 'timestamp']

userId	é categórico nominal,	identificador de cada usuário.
movieId	é categórico nominal,	identificador do filme.
rating	é categórico ordinal, avaliação de 0.5 a 5.0 em passos de 0.5.
timestamp	é categórico ordinal, instante em que a nota foi dada.
```

```python
tags.columns = ['userId', 'movieId', 'tag', 'timestamp']

userId	é categórico nominal,	Identificador do usuário.
movieId	é categórico nominal,	Identificador do filme.
tag	categórico nominal, 	palavra ou frase associada ao filme.
timestamp	é categórico ordinal, instante em que a tag foi dada.
```

```python
links.columns = ['movieId', 'imdbId', 'tmdbId']

movieId	é categórico nominal,	identificador do filme.
imdbId	é categórico nominal, do mesmo filme no IMDb.
tmdbId	é categórico nominal, do mesmo filme no TMDb.
```
