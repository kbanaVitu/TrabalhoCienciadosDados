Date é uma feature do tipo data(ou date em ingles) que significa data de dia, sendo categórica ordinal
   
Time é uma feature categorica que significa tempo, sendo categórica ordinal

CO(GT) é uma feature númerica que significa: Concentração média horária real de CO em mg/m^3, numérica contínua

PT08.S1(CO) é uma feature numérica contínua que significa: Resposta média horária do sensor (nominalmente CO alvo)

NMHC(GT) é uma feature númerica contínua que significa: Concentração média horária real de hidrocarbonetos não metânicos em microg/m^3

C6H6(GT) é uma feature numérica continua que significa: Concentração média horária real de benzeno em microg/m^3

PT08.S2(NMHC)	é uma feature numérica contínua que significa: resposta média do sensor por hora (nominalmente direcionada ao NMHC)

NOx(GT) é uma feature númerica contínua que significa: Concentração média horária real de NOx em ppb

PT08.S3(NOx) é uma feature numérica contínua que significa: resposta média do sensor por hora (nominalmente NOx direcionado)

NO2(GT) é uma feature númerica contínua que significa: Concentração média horária real de NO2 em microg/m^3

PT08.S4(NO2) é uma feature numérica contínua que significa: resposta média do sensor por hora (nominalmente NO2 direcionado)

PT08.S5(O3) é uma feature numérica contínua que significa: resposta média do sensor por hora (nominalmente O3 direcionado)

T é uma feature numérica continua que significa temperatura

RH é uma feature numérica continua que significa umidade relativa

AH é uma feature numérica continua que significa umidade absoluta
